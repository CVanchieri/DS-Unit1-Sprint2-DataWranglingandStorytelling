## Simple Web Scraper with IMDb

```python
# 'import' 'get' to import the html link.
from requests import get
# label the html link file.
url = 'https://www.imdb.com/title/tt6105098/ratings?ref_=tt_ov_rt'
# use 'get' to retrieve the html file.
response = get(url)
# show 500 text count of the file.
print(response.text[:500])
```

```python
# we can use 'BeautfiulSoup' to use 'html.parser'.
from bs4 import BeautifulSoup
# create the new file using the 'BeautifulSoup'.
html_soup = BeautifulSoup(response.text, 'html.parser')
# show the new file in 'text'.
type(html_soup)
```

```python
# we can put all the 'votes' we want into 1 container.
vote_container = html_soup.find_all('div', class_ ='leftAligned')
# container size is 1-11.
vote_container[1:11]
```

```python
# now we can pull out just the vote count #.
votes = [containers.text for containers in vote_container][1:11]
# show all the votes.
votes
```

```python
# now we can scrape all these movies at once with a 'votes_list' function.
urls = ['https://www.imdb.com/title/tt6105098/ratings?ref_=tt_ov_rt',     # Lion King (2019)
        'https://www.imdb.com/title/tt0110357/ratings?ref_=tt_ov_rt',     # Lion King (1994)
        'https://www.imdb.com/title/tt6139732/ratings?ref_=tt_ov_rt',     # Aladdin (2019)
        'https://www.imdb.com/title/tt0103639/ratings?ref_=tt_ov_rt']     # Aladdin (1992) w/ Robin Williams


# 'votes_list' function that we can run through each movie url.
votes_list = []
for url in urls:
    # Get raw HTML response
    response = get(url)
    # Convert to BS Object
    html_soup = BeautifulSoup(response.text, 'html.parser')
    # Find vote containers and extract star ratings
    vote_containers = html_soup.find_all('div', class_ = 'leftAligned')
    votes = [containers.text for containers in vote_containers][1:11]
    # Append to initial list
    votes_list.append(votes)
    # show all the movies with all the corresponding votes.    
print(votes_list)
```

```python
# import pandas to create a data frame.
import pandas as pd
# label the movies to use as column headers in the data set.
movies = ['The Lion King (2019)', 'The Lion King (1994)', 'Aladdin (2019)', 'Aladdin (1992)']
# create a data frame from the 'votes_list' we created above.
df = pd.DataFrame(votes_list)
# create the '.T' transverse matrix.
df = df.T
# set the data set columns to the 'movies' from above.
df.columns = movies
# we can remove any ',' in the data frame.
df = df.apply(lambda x: x.str.replace(',',''))
# create an end column with the 'Star Rating'.
df['Star Rating'] = range(1,11)[::-1]
# show the data set with headers.
df
```

```python
df
```

```python
# Convert df into tidy-format
# 'melt' on 'Star Rating'.
df_tidy = df.melt(id_vars='Star Rating')

# we can rename the variable & value columns
df_tidy = df_tidy.rename(columns={'variable': 'Movie', 'value': 'Number of Votes'})
# 'Number of Votes' is an object but we can set it to an int withpd.to_numeric.
df_tidy['Number of Votes'] = pd.to_numeric(df_tidy['Number of Votes'])
#show the data frame with headers.
df_tidy.head()
```

```python
# create a end column where we will show the % of the 'Number of Votes'.
df_tidy['Vote Percent'] = df_tidy.groupby('Movie')['Number of Votes'].apply(lambda x: x / x.sum() * 100)
# show the data frame wit hthe new column.
df_tidy.head()
```

```python
# we can check that all values are usable for charting.
df_tidy.info()
```

```python
# import seaborn for charts.
import seaborn as sns
# use catplot so we can chart each movie seperately in a bar graph.
sns.catplot(x='Star Rating', y='Vote Percent', col='Movie', col_wrap=2, height=6, kind='bar', data=df_tidy);
```

## Reproduce with real data

```python
# import he data set .csv file.
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
# show the shape of the data.
print(df.shape)
# show the data frame with headers.
df.head()
```

```python
# look at all the value counts for the 'category' column.
df.category.value_counts()
```

```python
# look at the data types for each column.
df.dtypes
```

```python
# use pd.to_datetime() on 'timestamp' so it is readable.
df['timestamp'] = pd.to_datetime(df['timestamp'])

# look at the statistics for 'timestamp' after using pd.to_datetime.
df.timestamp.describe()
```

```python
# we are creating a data set of just the 'IMDb users' from the 'category' column.
df_imdb = df[df.category == 'IMDb users']
# show the data set shape.
df_imdb.shape
```

```python
# show the data set tail end, this one should have all the info we need.
final = df_imdb.tail(1)

# show the final row.
final
```

```python
# we can pull together all the _pct columns.
#columns = ['{}_pct'.format(i) for i in range(1,11)]
columns = [f'{i}_pct' for i in range(1,11)]

# show the columns list.
columns
```

```python
# we now use the 'final' data set and keep only the _pct 'columns' we set.
data = final[columns]

# create the .T transverse data set.
data = data.T

# set the index range (rows).
data.index = range(1,11)

# show the data set.
data
```

```python
# use the 'fivethirtyeight' style.
plt.style.use('fivethirtyeight')

# create the bar plot with the color, width, and remove the legend.
data.plot.bar(color='C1', width=0.9, legend=False)

# add the main title text.
plt.text(x=-1.5, y=50, fontsize=16, fontweight='bold',
         s = "'An Inconvenient Sequel: Truth to Power' is divisive")

# add the subtitle text.
plt.text(x=-1.5, y=46, fontsize=12,
         s = "IMDb ratings for the film as of Aug. 29")

# add the y axis tick marks.
plt.yticks([0, 10, 20, 30, 40])

# add the text to the x label.
plt.xlabel('Rating', fontsize=10, fontweight='bold')

# add the text to the y label.
plt.ylabel('Percent of Total Votes', fontsize=10, fontweight='bold');
```

## Quick Introduction to Altair Plotting Package

```python
import altair as alt
from vega_datasets import data

source = data.cars()
source.head()
```

```python
brush = alt.selection(type='interval', resolve='global')

base = alt.Chart(source).mark_point().encode(
    y='Miles_per_Gallon',
    color=alt.condition(brush, 'Origin', alt.ColorValue('gray')),
    tooltip=['Name','Origin','Weight_in_lbs'],
).add_selection(
    brush
).properties(
    width=350,
    height=350
)

base.encode(x='Horsepower') | base.encode(x='Acceleration') | base.encode(x='Weight_in_lbs')
```
